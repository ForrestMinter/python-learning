```python
# 作业1：判断奇数
print("=" * 40)
print("作业1：判断奇数")
print("=" * 40)

def isOdd(num):
    return num % 2 == 1

try:
    user_input = int(input("请输入一个整数："))
    if isOdd(user_input):
        print(f"{user_input} 是奇数")
    else:
        print(f"{user_input} 是偶数")
except ValueError:
    print("输入错误！请输入一个整数。")
```

```python
# 作业2：判断字符串是否为数字表示
print("\n" + "=" * 40)
print("作业2：判断字符串是否为数字表示")
print("=" * 40)

def isNum(s):
    try:
        complex(s)
        return True
    except ValueError:
        return False

user_input = input("请输入一个字符串：")
if isNum(user_input):
    print(f"{user_input} 是数字表示")
else:                
    print(f"{user_input} 不是数字表示")
```

```python
# 作业3：判断质数
print("\n" + "=" * 40)
print("作业3：判断质数")
print("=" * 40)

def isPrime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
try:
    num = int(input("请输入一个整数："))
    if isPrime(num):
        print(f"{num} 是质数")
    else:
        print(f"{num} 不是质数")
except ValueError:
    print("输入错误！请输入一个整数。")
```

```python
# 作业4：小学生加减法练习器
import random

print("\n" + "=" * 40)
print("作业4: 小学生加减法练习器")
print("=" * 40)

## (1)
print("\n(1) 随机生成一道加法题：")
a = random.randint(0, 10)
b = random.randint(0, 10)
print(f"  {a} + {b} = ?")

## (2)
print("\n(2) 函数生成加法题并输出：")
def opera2():
    a = random.randint(0, 10)
    b = random.randint(0, 10)
    print(f"{a} + {b} = ?")
    return a, b
opera2()

## (3)
print("\n(3) 函数返回加数，主模块输出：")
def opera3():
    a = random.randint(0, 10)
    b = random.randint(0, 10)
    return a, b
a, b = opera3()
print(f"{a} + {b} = ?")

## (4)
print("\n(4) 生成10道加法题：")
def create1(n):
    questions = []
    for _ in range(n):
        a = random.randint(0, 10)
        b = random.randint(0, 10)
        questions.append((a, b))
    return questions

questions1 = create1(10)
for a, b in questions1:
    print(f"  {a} + {b} = ?")

## (5)
print("\n(5) 生成10道加减法题：")
def create2(n):
    questions = []
    for _ in range(n):
        a = random.randint(0, 10)
        b = random.randint(0, 10)
        op = random.choice(['+', '-'])
    
        if op == '-' and a < b:
            a, b = b, a
        questions.append((a, b, op))
    return questions

questions2 = create2(10)
for a, b, op in questions2:
    print(f"  {a} {op} {b} = ?")
```

```python
# 作业5：乘法计算器
def multi(*args):
    result = 1
    for num in args:
        result *= num
    return result

print("\n" + "=" * 40)
print("作业5: 乘法计算器")
print("=" * 40)

while True:
    user_input = input("请输入数字（用空格分隔）：")
    
    try:
        numbers = [float(x) for x in user_input.split()]
        
        if len(numbers) == 0:
            print("请至少输入一个数字！")
            continue
        product = multi(*numbers)
        
        print(f"乘积：{product}")
        
    except ValueError:
        print("输入错误！请输入数字，用空格分隔。")
    print("-" * 40)
```

```python
# 作业6：作物生长阶段函数嵌套调用
print("\n" + "=" * 40)
print("作业6: 作物产量预测")
print("=" * 40)

## (1)
def growth_stage(day):
    if day < 30:
        return "幼苗期"
    elif day < 60:
        return "拔节期"
    elif day < 90:
        return "抽穗期"
    else:
        return "成熟期"

## (2) 
def yield_predict(stage, area):
    stage_yield = {
        "幼苗期": 0,
        "拔节期": 200,
        "抽穗期": 500,
        "成熟期": 800
    }
    return stage_yield[stage] * area

# (3) 主函数，调用上述函数计算总产量
def main(day, area):
    if day < 0 or area < 0:
        return 0
    
    stage = growth_stage(day)
    total_yield = yield_predict(stage, area)
    return total_yield

print("\n作物生长天数与产量预测")
print("-" * 40)

while True:
    try:
        day = float(input("请输入生长天数："))
        area = float(input("请输入种植面积（亩）："))
        
        if day < 0 or area < 0:
            print("输入无效！天数和面积不能为负数。")
            continue
        
        result = main(day, area)
        stage = growth_stage(day)
        
        print(f"\n生长天数：{day}天")
        print(f"作物阶段：{stage}")
        print(f"种植面积：{area}亩")
        print(f"预测总产量：{result}kg")
        
    except ValueError:
        print("输入错误！请输入数字。")
    
    print("-" * 40)
    again = input("是否继续？(y/n)：")
    if again.lower() != 'y':
        print("程序结束")
        break
    print()
```

```python
# 作业7：pH值分析器
print("\n" + "=" * 40)
print("作业7: pH值分析器")
print("=" * 40)

def ph_analyzer(*readings):
    if len(readings) < 3:
        return "数据不足"
    
    avg = sum(readings) / len(readings)

    if avg < 6.5:
        status = "酸性土壤"
    elif avg > 7.5:
        status = "碱性土壤"
    else:
        status = "中性土壤"

    return round(avg, 2), status

print("\n土壤pH值分析")
print("输入至少3个pH值，用空格分隔")
print("-" * 40)

while True:
    user_input = input("请输入pH值：")
    
    try:
        readings = [float(x) for x in user_input.split()]
        
        result = ph_analyzer(*readings)
        
        if result == "数据不足":
            print(f"结果：{result}（至少需要3个检测点）")
        else:
            avg, status = result
            print(f"pH值数据：{readings}")
            print(f"平均值：{avg}")
            print(f"土壤状态：{status}")
            
    except ValueError:
        print("输入错误！请输入数字，用空格分隔。")
    
    print("-" * 40)
```

```python
# 作业8：病虫害预警函数
print("\n" + "=" * 40)
print("作业8: 病虫害预警系统")
print("=" * 40)

def pest_alert(**sensor_data):
    temperature = sensor_data.get('temperature', 25)
    humidity = sensor_data.get('humidity', 60)
    infection = sensor_data.get('infection', False)

    if temperature > 30 and humidity > 80:
        return "高温高湿预警"
    elif infection and temperature > 25:
        return "病害扩散预警"
    else:
        return "环境正常"

print("\n病虫害预警系统")
print("输入温度和湿度，可选项输入是否已有病害")
print("-" * 40)

while True:
    user_input = input("请输入温度和湿度（格式：温度 湿度）：")
    
    if user_input.lower() == 'q':
        print("程序结束")
        break
    
    try:
        parts = user_input.split()
        if len(parts) < 2:
            print("输入格式错误！请输入温度和湿度，用空格分隔。")
            continue
        
        temp = float(parts[0])
        hum = float(parts[1])
        
        infection_input = input("是否有病害？(y/n，默认n)：")
        has_infection = infection_input.lower() == 'y'
        
        result = pest_alert(temperature=temp, humidity=hum, infection=has_infection)
        
        print(f"\n温度：{temp}℃")
        print(f"湿度：{hum}%")
        print(f"已有病害：{'是' if has_infection else '否'}")
        print(f"预警结果：{result}")
        
    except ValueError:
        print("输入错误！温度和湿度必须是数字。")
    
    print("-" * 40)
```
