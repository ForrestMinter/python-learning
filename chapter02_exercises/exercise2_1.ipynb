```python
# 作业1：温度转换
print("=" * 40)
print("作业1：温度转换")
print("=" * 40)

TempStr = input("请输入带有符号的温度值（如 32F 或 100C）：")
if TempStr[-1] in ['F', 'f']:
    C = (eval(TempStr[0:-1]) - 32) / 1.8
    print("转换后的温度是：{:.2f}C".format(C))
elif TempStr[-1] in ['C', 'c']:
    F = 1.8 * eval(TempStr[0:-1]) + 32
    print("转换后的温度是：{:.2f}F".format(F))
else:
    print("输入格式错误")
```

```python
# 作业2：改造温度转换
print("\n" + "=" * 40)
print("作业2：改造温度转换")
print("=" * 40)

C = eval(input("请输入摄氏温度（只输入数字）："))
F = int(1.8 * C + 32)
print("转换后的华氏温度是：{}°F".format(F))
```

```python
# 作业3：汇率兑换
print("\n" + "=" * 40)
print("作业3：汇率兑换")
print("=" * 40)

rate = 7.0
money = input("请输入金额及货币符号（如100R或50D）：")
#禁用客户的¥字符，避免¥被当作多个字节处理导致判断失败

if money[-1] in ['R', 'r']:
    dollar = eval(money[0:-1]) / rate
    print("兑换后的美元为：{:.2f}$".format(dollar))
elif money[-1] in ['D', 'd', '$']:
    rmb = eval(money[0:-1]) * rate
    print("兑换后的人民币为：{:.2f}¥".format(rmb))
else:
    print("输入格式错误")
```

```python
# 作业4：八进制转换
print("\n" + "=" * 40)
print("作业4: 八进制转换")
print("=" * 40)

while True:
    oct_str = input("请输入一个八进制数（如123，只能包含0-7）：")
    
    try:
        dec_num = int(oct_str, 8)
        break  
    except ValueError:
        print("输入错误，八进制只能包含数字0-7，请重新输入。")
        print() 

print("\n转换成功！")
print("-" * 20)
print("十进制数为：", dec_num)
print("二进制数为：", bin(dec_num))
print("十六进制数为：", hex(dec_num))
```

```python
# 作业6：利用math库运行语句并打印
print("\n" + "=" * 40)
print("作业6：利用math库运行语句并打印")        
print("=" * 40)

import math

print("1.", math.sin(2 * math.pi))
print("2.", math.floor(-2.5))
print("3.", math.ceil(3.5 + math.floor(-2.5)))
print("4.", round(math.fabs(-2.5)))
print("5.", math.sqrt(math.pow(2, 4)))
print("6.", math.log(math.e))
print("7.", math.gcd(12, 9))
print("8.", math.fmod(36, 5))
```

```python
# 作业7：体重计算
print("\n" + "=" * 40)
print("作业7: 体重计算")
print("=" * 40)

earth_weight = eval(input("请输入你现在的体重（kg）："))

print("\n年份\t地球体重(kg)\t月球体重(kg)")
print("-" * 40)

for year in range(1, 11):
    earth_weight += 0.5
    moon_weight = earth_weight * 0.165
    print(f"{year}\t{earth_weight:.1f}\t\t{moon_weight:.2f}")

print("=" * 40)
```
