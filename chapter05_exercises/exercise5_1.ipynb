```python
# 作业1:月份字典
print("=" * 40)
print("作业1:月份字典")
print("=" * 40)

month_dict = {
    'Jan': 31, 'Feb': 28, 'Mar': 31, 'Apr': 30, 'May': 31, 'Jun': 30,
    'Jul': 31, 'Aug': 31, 'Sep': 30, 'Oct': 31, 'Thi': 40
}   
print("原始字典：", month_dict)

##（1):追加新的键值对
x = {'Nov': 30, 'Dec': 31}
month_dict.update(x)
print("追加后的字典：", month_dict)

##（2):删除键值对
del month_dict['Thi']
print("删除后的字典：", month_dict)

##（3):修改键值对
month_dict['Feb'] = 29
print("修改后的字典：", month_dict)

##（4):查询键值对
print("查询Oct对应的值：", month_dict['Oct'])

## (5):查询键值对
if 'Xyz' in month_dict:
    print("查询Xyz对应的值：", month_dict['Xyz'])
else:
    print("No Found")

## (6):输出所有键
print("所有键：", month_dict.keys())

## (7):输出所有值
print("所有值：", month_dict.values())

## (8):输出所有键值对
print("所有键值对：", month_dict.items())
```

```python
# 作业2:姓名成绩字典
print("\n" + "=" * 40)
print("作业2:姓名成绩字典")
print("=" * 40)

score_dict = {
    "王星": 87,
    "刘艺": 77,
    "马跃": 60,
    "李小辉": 94,
    "刘天天": 95,
    "秦天": 85
}
print("原始字典：", score_dict)
average_score = sum(score_dict.values()) / len(score_dict)
print("平均成绩：", average_score)

max_name = max(score_dict, key=score_dict.get)
print(f"最高分学生：{max_name}")
```

```python
# 作业3:根据学号查询姓名
print("\n" + "=" * 40)
print("作业3:学号查询姓名")
print("=" * 40)

dic = {'Wangbing': 1001, 'Malins': 1003, 'Xulei': 1004}
reverse_dic = {v: k for k, v in dic.items()}

num = int(input("请输入学号："))

if num in reverse_dic:
    print(reverse_dic[num])
else:
    print("No Found")
```

```python
# 作业4:判断字符串是否含有元音字母
print("\n" + "=" * 40)
print("作业4:判断元音字母")
print("=" * 40)

vowels = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
s = input("请输入字符串：")

has_vowel = False
for ch in s:
    if ch in vowels:
        has_vowel = True
        break

print(has_vowel)
```

```python
# 作业5:集合运算
print("\n" + "=" * 40)
print("作业5:集合运算")
print("=" * 40)

classA = {1, 3, 7, 15, 22, 26, 27}
classB = {1, 4, 8, 9, 15, 20, 26, 30}
library = {1, 8, 15, 22, 29}

print(f"classA: {classA}")
print(f"classB: {classB}")
print(f"library: {library}")
print("-" * 40)

## (1):两个班级都想去
result1 = classA & classB
print(f"(1) 两个班级都想去: {result1}")

## (2):仅有一个班级想去
result2 = classA ^ classB
print(f"(2) 仅有一个班级想去: {result2}")

## (3):A班想去但B班不想去
result3 = classA - classB
print(f"(3) A想去B不想去: {result3}")

## (4):两个班级能一起去
result4 = classA & classB & library
print(f"(4) 两个班级都能去: {result4}")

## (5):A班能去而B班不能去
result5 = classA & library - classB
print(f"(5) A能去B不能去: {result5}")
```
