```python
# 作业1：斐波那契数列（递归）
print("=" * 40)
print("作业1: 斐波那契数列")
print("=" * 40)

def fib(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)

n = int(input("请输入n："))
print(f"fib({n}) = {fib(n)}")
```

```python
# 作业2：角谷猜想（循环版）
print("=" * 40)
print("作业2: 角谷猜想（循环）")
print("\n" + "=" * 40)

def jiaogu_loop(n):
    while n != 1:
        if n % 2 == 0:
            print(f"{n}/2={n//2}", end=" ")
            n = n // 2
        else:
            print(f"{n}*3+1={n*3+1}", end=" ")
            n = n * 3 + 1
    print(n)

num = int(input("请输入自然数："))
jiaogu_loop(num)
```

```python
# 作业2：角谷猜想（递归版）
print("=" * 40)
print("作业2: 角谷猜想（递归）")
print("\n" + "=" * 40)

def jiaogu_recursive(n):
    print(n, end=" ")
    if n == 1:
        return
    if n % 2 == 0:
        jiaogu_recursive(n // 2)
    else:
        jiaogu_recursive(n * 3 + 1)

num = int(input("请输入自然数："))
jiaogu_recursive(num)
print()
```

```python
# 作业3：人员信息管理系统
print("=" * 40)
print("作业3: 人员信息管理系统")
print("\n" + "=" * 40)

person_info = {
    'Zhangsan': {'phone': 111111, 'addr': 'JS'},
    'Lisi': {'phone': 222222, 'addr': 'SH'}
}

def addInfo(info_dict):
    name = input("请输入需要添加的人名：")
    
    if name in info_dict:
        choice = input(f"人名{name}已存在，按1覆盖原来的信息，其他键放弃本次操作：")
        if choice != '1':
            print("操作已取消")
            return info_dict
    
    phone = input("请输入电话号码：")
    addr = input("请输入省份缩写：")
    info_dict[name] = {'phone': phone, 'addr': addr}
    print("信息成功添加/更新！")
    return info_dict

def search(info_dict):
    name = input("请输入需要查询的人名：")
    if name in info_dict:
        info = info_dict[name]
        print(f"{name}:{info['phone']}@{info['addr']}")
    else:
        print("not found")

def main():
    global person_info
    while True:
        print("\n请输入操作选项：")
        print("1. 添加  2. 查询  3. 退出")
        choice = input()
        
        if choice == '1':
            person_info = addInfo(person_info)
        elif choice == '2':
            search(person_info)
        elif choice == '3':
            print("欢迎下次使用，再见")
            break
        else:
            print("输入错误，请重新选择")

if __name__ == "__main__":
    main()
```

```python
# 作业5：递归作物生长
print("=" * 40)
print("作业5: 递归作物生长")
print("\n" + "=" * 40)

MAX_DAYS = 30

def grow(height, rate, day=0):
    if day > MAX_DAYS:
        raise ValueError(f"生长天数超过限制{MAX_DAYS}天")
    
    if day == 0:
        print(f"第{day}天：{height:.2f}米")
    
    if day == MAX_DAYS:
        return height
    
    new_height = height * rate
    print(f"第{day+1}天：{new_height:.2f}米")
    return grow(new_height, rate, day + 1)

try:
    result = grow(0.5, 1.05)
    print(f"\n30天后作物高度：{result:.2f}米")
except ValueError as e:
    print(f"异常：{e}")
```

```python
# 作业6：温度异常检测
print("=" * 40)
print("作业6: 温度异常检测")
print("\n" + "=" * 40)

celsius_temps = [22.3, 19.8, 25.6, 18.9]
print(f"摄氏度列表：{celsius_temps}")

# (1) 
c_to_f = lambda c: c * 9/5 + 32
fahrenheit_temps = [c_to_f(c) for c in celsius_temps]
print(f"华氏度列表：{fahrenheit_temps}")

# (2) 
def detect_abnormal(temperatures, deviation=3):
    """检测偏离平均值±deviation的温度"""
    avg = sum(temperatures) / len(temperatures)
    print(f"\n平均值：{avg:.2f}")
    
    abnormal = []
    for temp in temperatures:
        if abs(temp - avg) > deviation:
            abnormal.append(temp)
    
    return abnormal

abnormal_temps = detect_abnormal(celsius_temps, 3)
print(f"异常温度（偏离平均值>3°C）：{abnormal_temps}")
```

```python
# 作业7：农药喷洒时间调度
from datetime import datetime, timedelta

print("=" * 40)
print("作业7: 农药喷洒时间调度")
print("\n" + "=" * 40)

def schedule_spray(start_date, interval, times):
    start = datetime.strptime(start_date, "%Y-%m-%d")
    
    spray_dates = []
    for i in range(times):
        spray_date = start + timedelta(days=i * interval)
        spray_dates.append(spray_date.strftime("%Y-%m-%d"))
    
    return spray_dates

start = input("请输入起始日期（YYYY-MM-DD）：")
interval = int(input("请输入间隔天数："))
times = int(input("请输入喷洒次数："))

dates = schedule_spray(start, interval, times)
print(f"\n喷洒日期列表：{dates}")
```
