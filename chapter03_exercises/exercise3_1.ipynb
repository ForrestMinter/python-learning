```python
# 作业1:角度与弧度转换
print("=" * 40)
print("角度与弧度转换")
print("=" * 40)

import math

## (1):角度转弧度
deg = 47
rad = math.radians(deg) ### 等价于 deg * math.pi / 180
print(rad)

## (2):弧度转角度
rad = math.pi / 7
deg = math.degrees(rad) ### 等价于 rad * 180 / math.pi
print(deg)
```

```python
# 作业2:表达式计算
print("\n" + "=" * 40)
print("表达式计算")
print("=" * 40)

import math
x = float(input("请输入x的值："))
y = float(input("请输入y的值："))

## (1):
expr1 = math.exp(x) / ( (abs(x ** y - y ** x) + 1))
print(expr1)

## (2):
expr2 = (1 + math.sqrt(abs(x + y))) / (math.exp(x) * y)
print(expr2)
```

```python
# 作业3:字符串操作
print("\n" + "=" * 40)
print("字符串操作")
print("=" * 40)

s = "SUN YAT-SEN UNIVERSITY"

## (1):
print("(1)", s[0:3], s[4:11], s[12:])

## (2):
print("(2)", s[0] + s[4] + s[8] +s[12])

## (3):
char = input("(3) 请输入要统计的字符：")
count = s.count(char)
print("(3)", f"字符'{char}'出现了{count}次")

## (4):
print("(4)", s.split(' '))

## (5):
print("(5)", s.lower())
```

```python
# 作业4:回文数判断
print("\n" + "=" * 40)
print("回文数判断")
print("=" * 40)

num = input("请输入一个五位数：")
if num == num[::-1]:
    print(f"{num}：这个数是回文数")
else:
    print(f"{num}：这个数不是回文数")
```

```python
# 作业5:按空格分割字符串
print("\n" + "=" * 40)
print("按空格分割字符串")
print("=" * 40)

s = input("请输入一个字符串(单词间用空格分割）：")
words = s.split(' ')
print("分割后的单词列表：", words)
print("\n分割结果：")
for word in words:
    print(word)
```

```python
# 作业6:数字转星期
print("\n" + "=" * 40)
print("数字转星期")
print("=" * 40)
num = int(input("请输入一个数字(1-7)：")) ## int:将输入的内容（字符串）转换为整数
weekdays = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
if 1 <= num <= 7:
    print(f"{num}对应的星期是：{weekdays[num - 1]}")
else:
    print("输入错误，请输入1-7之间的数字。") 
```

```python
# 作业7:计算太阳高度角
print("\n" + "=" * 40)
print("计算太阳高度角")
print("=" * 40)

import math

phi = float(input("请输入当地纬度（度）："))
N = float(input("请输入距春分天数："))
t = float(input("请输入地方时（小时，0-24）："))

phi_rad = math.radians(phi)
delta_deg = 23.5 * math.sin(math.radians(N))
delta_rad = math.radians(delta_deg)

term = (math.sin(phi_rad) * math.sin(delta_rad) +
        math.cos(phi_rad) * math.cos(delta_rad) *
        math.cos(math.radians(15 * (t - 12))))


if -1 <= term <= 1:
    h_rad = math.asin(term)
    h_deg = math.degrees(h_rad)
    print(f"\n太阳高度角：{h_deg:.2f}°")
```
