```python
# 作业1：判断闰年
print("\n" + "=" * 40)
print("作业1: 判断闰年")
print("=" * 40)

year = int(input("请输入年份："))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year}年是闰年")
else:
    print(f"{year}年不是闰年")
```

```python
# 作业2：分段函数
print("\n" + "=" * 40)
print("作业2: 分段函数")
print("=" * 40)

x = float(input("请输入x的值："))

if x < 0:
    fx = x ** 2
elif x < 10:
    fx = 2 * x + 1
else:
    fx = 3 * x - 5

print(f"f({x}) = {fx}")
```

```python
# 作业3：舒适温度判断
print("\n" + "=" * 40)
print("作业3: 舒适温度判断")
print("=" * 40)

season = input("请输入季节（春夏秋冬）：")
temp = float(input("请输入温度："))

if season == "春" or season == "秋":
    if 18 <= temp <= 25:
        print("舒适")
    else:
        print("不舒适")
elif season == "夏":
    if 19 <= temp <= 24:
        print("舒适")
    else:
        print("不舒适")
elif season == "冬":
    if 12 <= temp <= 22:
        print("舒适")
    else:
        print("不舒适")
else:
    print("季节输入错误")
```

```python
# 作业4：计算中位数
print("\n" + "=" * 40)
print("作业4: 计算中位数")
print("=" * 40)

data = input("请输入数列（逗号分隔）：")
numbers = [int(x) for x in data.split(',')]
numbers.sort()

n = len(numbers)
if n % 2 == 1:
    median = numbers[n // 2]
else:
    median = (numbers[n // 2 - 1] + numbers[n // 2]) / 2

print(median)
```

```python
# 作业5：统计不同字符个数
print("\n" + "=" * 40)
print("作业5: 统计字符个数")
print("=" * 40)

s = input("请输入一行字符：")

letters = 0  
digits = 0    
spaces = 0    
others = 0    

for ch in s:
    if ch.isalpha():         
        letters += 1
    elif ch.isdigit():        
        digits += 1
    elif ch.isspace():       
        spaces += 1
    else:
        others += 1

print(f"英文字符: {letters}")
print(f"数字: {digits}")
print(f"空格: {spaces}")
print(f"其他字符: {others}")
```

```python
# 作业6：作物灌溉预警
print("\n" + "=" * 40)
print("作业6: 灌溉预警")
print("=" * 40)

moisture = float(input("请输入土壤湿度（%）："))

if moisture < 30:
    print("紧急灌溉")
elif moisture < 50:
    print("建议灌溉")
else:
    print("保持现状")
```

```python
# 作业7：病虫害风险预测
print("\n" + "=" * 40)
print("作业7: 风险预测")
print("=" * 40)

temperature = float(input("请输入温度（℃）："))
humidity = float(input("请输入湿度（%）："))

if 25 <= temperature <= 30 and humidity > 80:
    print("高风险")
else:
    print("低风险")
```

```python
# 作业8：产量达标统计
print("\n" + "=" * 40)
print("作业8: 产量达标统计")
print("=" * 40)

yields = [int(x) for x in input("请输入各地块产量（逗号分隔）：").split(',')]

count = 0
for y in yields:
    if y >= 500:
        count += 1

print(f"达标地块数：{count}")
```

```python
# 作业9：高温天数占比
print("\n" + "=" * 40)
print("作业9: 高温天数占比")
print("=" * 40)

temps = [int(x) for x in input("请输入温度数据（逗号分隔）：").split(',')]

hot_days = 0
for t in temps:
    if t > 30:
        hot_days += 1

ratio = hot_days / len(temps) * 100
print(f"高温比例：{ratio:.1f}%")
```

```python
# 作业10：温室环境调控
print("\n" + "=" * 40)
print("作业10: 温室调控")
print("=" * 40)

data_input = input("请输入温湿度数据（如(26,80),(29,76)）：")
data = eval(f"[{data_input}]")

for i, (temp, humidity) in enumerate(data):
    if temp > 28 and humidity > 75:
        print(f"第{i+1}小时需通风")
```

```python
# 作业11：异常坐标检测
print("\n" + "=" * 40)
print("作业11: 异常坐标检测")
print("=" * 40)

ndvi_matrix = [
    [0.3, 0.1, 0.4],
    [0.15, 0.25, 0.18]
]

for i in range(len(ndvi_matrix)):
    for j in range(len(ndvi_matrix[i])):
        if ndvi_matrix[i][j] < 0.2:
            print(f"异常坐标：({i},{j})")
```
