```python
# 作业1
from PIL import Image, ImageEnhance, ImageFilter

# 打开原始图像
img = Image.open("input.jpg")

# (1) 生成缩略图
thumb = img.copy()
thumb.thumbnail((128, 128))
thumb.save("1_thumbnail.jpg")

# (2) 截取子图
box = (50, 50, 200, 200) 
crop = img.crop(box)
crop.save("2_crop.jpg")

# (3) 获取图像尺寸并缩小一半
width, height = img.size
resized = img.resize((width // 2, height // 2))
resized.save("3_half_size.jpg")

# (4) 逆时针旋转45度
rotated = img.rotate(45, expand=True)  # expand=True 保证全部内容可见
rotated.save("4_rotate45.jpg")

# (5) 左右翻转
flipped = img.transpose(Image.FLIP_LEFT_RIGHT)
flipped.save("5_flip_lr.jpg")

# (6) 交换红波段（R）和蓝波段（B）
r, g, b = img.split()
swapped = Image.merge("RGB", (b, g, r))
swapped.save("6_swap_rb.jpg")

# (7) 生成图像轮廓
contour = img.filter(ImageFilter.CONTOUR)
contour.save("7_contour.jpg")

# (8) 亮度增强2倍
enhancer = ImageEnhance.Brightness(img)
bright = enhancer.enhance(2.0)
bright.save("8_brightness.jpg")

# (9) 对比度增强2倍
contrast_enhancer = ImageEnhance.Contrast(img)
contrast = contrast_enhancer.enhance(2.0)
contrast.save("9_contrast.jpg")

# (10) 边界加强
edge = img.filter(ImageFilter.EDGE_ENHANCE)
edge.save("10_edge_enhance.jpg")

# (11) 锐化
sharp = img.filter(ImageFilter.SHARPEN)
sharp.save("11_sharpen.jpg")

# (12) 浮雕效果
emboss = img.filter(ImageFilter.EMBOSS)
emboss.save("12_emboss.jpg")

# (13) 平滑效果
smooth = img.filter(ImageFilter.SMOOTH)
smooth.save("13_smooth.jpg")
```

```python
# 作业2
from PIL import Image
import numpy as np

img = Image.open("input.jpg")
r, g, b = img.split()

# 转换为 numpy 数组方便运算
r_arr = np.array(r, dtype=np.float32)
g_arr = np.array(g, dtype=np.float32)
b_arr = np.array(b, dtype=np.float32)

# 红波段： *0.8 + 10
r_new = r_arr * 0.8 + 10
r_new = np.clip(r_new, 0, 255).astype(np.uint8)

# 蓝波段： *1.2
b_new = b_arr * 1.2
b_new = np.clip(b_new, 0, 255).astype(np.uint8)

# 绿波段保持不变
g_new = g_arr.astype(np.uint8)

# 合并
new_img = Image.merge("RGB", (Image.fromarray(r_new),
                               Image.fromarray(g_new),
                               Image.fromarray(b_new)))
new_img.save("2_band_processed.jpg")
new_img.show()
```

```python
# 作业3
from PIL import Image, ImageEnhance

gif = Image.open("animation.gif")
frame_index = 0

try:
    while True:
        frame = gif.copy()
        # 确保模式为 RGB
        if frame.mode != 'RGB':
            frame = frame.convert('RGB')
        enhancer = ImageEnhance.Brightness(frame)
        bright_frame = enhancer.enhance(1.5)
        bright_frame.save(f"frame_{frame_index}.png")
        frame_index += 1
        gif.seek(gif.tell() + 1)
except EOFError:
    print(f"处理完成，共 {frame_index} 帧")
    
```

```python
# 作业4
from PIL import Image

img = Image.open("input.jpg")
frames = []

for i in range(100):
    angle = i * 10
    rotated = img.rotate(angle, expand=True)
    rotated.save(f"rotate_{i:03d}.jpg")
    frames.append(rotated)


frames[0].save("rotating.gif",
               save_all=True,
               append_images=frames[1:],
               duration=50,          
               loop=0)               
```

```python
# 作业5
def judge(x):
  
    seen = set()
    seq = []
    n = x
    while n != 1:
        if n in seen:         
            return []
        seen.add(n)
        seq.append(n)
        total = 0
        for ch in str(n):
            total += int(ch) ** 2
        n = total
    seq.append(1)             
    return seq

with open("result.txt", "w", encoding="utf-8") as f:
    for num in range(2, 101):
        path = judge(num)
        if path:
            line = "->".join(str(v) for v in path)
            f.write(line + "\n")

print("结果已写入 result.txt")
```
