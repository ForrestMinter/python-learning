```python
# 作业2：比较两个文件内容
import os

print("=" * 40)
print("作业2: 比较文件内容")
print("=" * 40)

def compare_files(file1, file2):
    if not os.path.exists(file1) or not os.path.exists(file2):
        return "文件不存在"
    
    if os.path.isdir(file1) or os.path.isdir(file2):
        return "输入的是文件夹路径，请输入文件路径"
    
    # 用二进制模式打开，不关心编码
    with open(file1, 'rb') as f1, open(file2, 'rb') as f2:
        content1 = f1.read()
        content2 = f2.read()
    
    if content1 == content2:
        return "OK!"
    else:
        return "NO!"

file1 = input("请输入第一个文件路径：")
file2 = input("请输入第二个文件路径：")

result = compare_files(file1, file2)
print(result)
```

```python
# 作业3：IPv4地址分类
import os

print("\n" + "=" * 40)
print("作业3: IPv4地址分类")
print("=" * 40)

def is_valid_ipv4(ip):
    """判断是否为有效的IPv4地址"""
    parts = ip.split('.')
    if len(parts) != 4:
        return False
    for part in parts:
        if not part.isdigit():
            return False
        num = int(part)
        if num < 0 or num > 255:
            return False
    return True

def get_ip_class(ip):
    """获取IP地址的类别"""
    first_num = int(ip.split('.')[0])
    if 1 <= first_num <= 127:
        return 'A'
    elif 128 <= first_num <= 191:
        return 'B'
    elif 192 <= first_num <= 255:
        return 'C'
    else:
        return None

# 读取文件
filename = input("请输入IP地址文件名：")

# 检查文件是否存在
if not os.path.exists(filename):
    print("文件不存在！")
    exit()

# 读取文件内容
with open(filename, 'r', encoding='utf-8') as f:
    content = f.read()

# 分割IP地址（按空格分隔）
ip_list = content.split()

# 分类统计
classes = {'A': [], 'B': [], 'C': []}

for ip in ip_list:
    if is_valid_ipv4(ip):
        ip_class = get_ip_class(ip)
        if ip_class:
            classes[ip_class].append(ip)

# 按各类地址数量降序排列
sorted_classes = sorted(classes.items(), key=lambda x: len(x[1]), reverse=True)

for cls, ips in sorted_classes:
    print(f"{cls}类地址{len(ips)}个：{', '.join(ips)}")
```

```python
# 作业4：读取数据文件并计算平均值
import os

print("=" * 40)
print("作业4: 数据文件处理")
print("=" * 40)

def read_average(file_path):
    """读取文件中的数字，计算平均值"""
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
        
        # 提取所有数字
        numbers = []
        for line in content.split('\n'):
            parts = line.replace(',', ' ').split()
            for part in parts:
                try:
                    numbers.append(float(part))
                except ValueError:
                    pass
        
        if numbers:
            return sum(numbers) / len(numbers)
        return 0
    except Exception as e:
        return 0

def scan_folder(folder_path):
    """递归扫描文件夹，找出所有txt文件"""
    all_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.txt') or file.endswith('.data'):
                all_files.append(os.path.join(root, file))
    return all_files

# 获取文件夹路径
folder_path = input("请输入数据文件夹路径：")

if not os.path.exists(folder_path):
    print("文件夹不存在！")
    exit()

# 递归扫描所有txt文件
file_paths = scan_folder(folder_path)

if not file_paths:
    print("没有找到任何数据文件！")
    exit()

file_names = []
averages = []

for file_path in file_paths:
    # 提取文件名（不含扩展名）
    filename = os.path.basename(file_path)
    name_without_ext = os.path.splitext(filename)[0]
    file_names.append(name_without_ext)
    
    avg = read_average(file_path)
    averages.append(avg)
    
    print(f"{name_without_ext} 的平均值是 {avg:.2f}")

print("\n" + "=" * 40)
print("文件名称列表:", file_names)
print("平均值列表:", [round(avg, 2) for avg in averages])
```
