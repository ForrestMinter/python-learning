```python
# 作业1：土壤pH调节
print("\n" + "=" * 40)
print("作业1: 土壤pH调节")
print("=" * 40)

current_ph = float(input("请输入初始pH值："))
count = 0

while current_ph < 6.0 or current_ph > 7.0:
    current_ph += 0.5
    count += 1

print(f"需调整{count}次")
```

```python
# 作业2：农药残留降解
print("\n" + "=" * 40)
print("作业2: 农药残留降解")
print("=" * 40)

residue = float(input("请输入初始残留量（mg/kg）："))
days = 0

while residue >= 1.0:
    residue *= 0.95   # 每天降解5%，剩余95%
    days += 1

print(f"达标天数：{days}")
```

```python
# 作业3：猜数游戏
import random

print("\n" + "=" * 40)
print("作业3: 猜数游戏")
print("=" * 40)

secret = random.randint(0, 9)
count = 0

while True:
    guess = int(input("请输入你猜的数（0-9）："))
    count += 1
    
    if guess > secret:
        print("太大了")
    elif guess < secret:
        print("太小了")
    else:
        print(f"预测{count}次，你猜中了！")
        break
```

```python
# 作业4：猜数游戏（异常处理版）
import random

print("\n" + "=" * 40)
print("作业4: 猜数游戏（异常处理）")
print("=" * 40)

secret = random.randint(0, 9)
count = 0

while True:
    try:
        guess = int(input("请输入你猜的数（0-9）："))
        count += 1
        
        if guess > secret:
            print("太大了")
        elif guess < secret:
            print("太小了")
        else:
            print(f"预测{count}次，你猜中了！")
            break
    except ValueError:
        print("输入错误！请输入一个整数（0-9）")
```

```python
# 作业5：生成固定随机数
import random

print("\n" + "=" * 40)
print("作业5: 生成固定随机数")
print("=" * 40)

random.seed(92)

## (1)
int_list = [random.randint(20, 50) for _ in range(20)]
print("(1) 整数列表:", int_list)

## (2) 
float_list = [random.uniform(20, 50) for _ in range(20)]
print("(2) 浮点数列表:", [round(x, 2) for x in float_list])

## (3)
random.shuffle(int_list)
print("(3) 打乱后:", int_list)

## (4)
sample_list = random.sample(int_list, 5)
print("(4) 随机取5个元素:", sample_list)
```

```python
# 作业6：作物生长周期模拟
import random

print("\n" + "=" * 40)
print("作业6: 作物生长周期模拟")
print("=" * 40)

random.seed()
consecutive_hot = 0
alert_triggered = False

for day in range(1, 31):
    temp = random.randint(15, 35)
    print(f"第{day}天: {temp}℃")
    
    if temp > 30:
        consecutive_hot += 1
        if consecutive_hot >= 3 and not alert_triggered:
            print("高温预警！作物可能受损")
            alert_triggered = True
    else:
        consecutive_hot = 0

if not alert_triggered:
    print("生长正常")
```

```python
# 作业7：降雨量统计
import random

print("\n" + "=" * 40)
print("作业7: 降雨量统计")
print("=" * 40)

random.seed()

light_rain = 0
moderate_rain = 0
heavy_rain = 0 

for _ in range(30):
    rain = random.randint(0, 100)
    if rain < 10:
        light_rain += 1
    elif rain <= 50:
        moderate_rain += 1
    else:
        heavy_rain += 1

print(f"小雨天数: {light_rain}")
print(f"中雨天数: {moderate_rain}")
print(f"大雨天数: {heavy_rain}")

if heavy_rain / 30 > 0.3:
    print("收成风险高")
else:
    print("收成稳定")
```

```python
# 作业8：病害风险预测
import random

print("\n" + "=" * 40)
print("作业8: 病害风险预测")
print("=" * 40)

random.seed()

while True:
    crop = input("请输入作物类型（小麦/玉米）：")
    
    if crop == "小麦" or crop == "玉米":
        break
    else:
        print("输入无效！请输入'小麦'或'玉米'")

probability = random.random()
print(f"病害发生概率: {probability:.2%}")

if probability > 0.7:
    print("高风险！建议防治")
else:
    print("低风险，继续观察")
```
