```python
# 作业3:统计单词出现次数
print("\n" + "=" * 40)
print("作业3:统计单词出现次数")
print("=" * 40)

sentence = input("请输入一个英文句子: ")
word = input("请输入要统计的单词: ")

sentence_clean = sentence.lower().replace('.', '').replace(',', '')
words = sentence_clean.split()

count_word = words.count(word.lower())
print(f"单词 '{word}' 在句子 '{sentence}' 中出现了 {count_word} 次")
```

```python
# 作业4:格式化输出作物信息
print("\n" + "=" * 40)
print("作业4:格式化输出作物信息")
print("=" * 40)

name = input("请输入作物名称：")
days = int(input("请输入生长周期（天）："))
temp = float(input("请输入适宜温度（℃）："))

output = "{:<10}|周期：{:>5d}天|温度{:>6.1f}℃".format(name, days, temp)

print(output)
```

```python
# 作业7:按绝对值大小输出列表
print("\n" + "=" * 40)
print("作业7:按绝对值大小输出列表")
print("=" * 40)

data = input("请输入列表数据（数字之间用逗号分隔）：")

numbers = [int(x) for x in data.split(',')]
sorted_numbers = sorted(numbers, key=abs, reverse=True)

print(sorted_numbers)
```

```python
# 作业8：统计不及格和优秀学生的平均成绩
print("\n" + "=" * 40)
print("作业8: 成绩统计")
print("=" * 40)

data = input("请输入成绩（用逗号分隔）：")

scores = [int(x) for x in data.split(',')]
print(f"输入的成绩：{scores}")

fail = []   #创建空列表用于存储不及格的成绩  
excellent = [] 

for score in scores:
    if score < 60:
        fail.append(score)
        print(f"{score}分 → 不及格")
    elif score >= 85:
        excellent.append(score)
        print(f"{score}分 → 优秀")
    else:
        print(f"{score}分 → 中等（不统计）")

print("-" * 40)


fail_count = len(fail)
if fail_count > 0:
    fail_avg = sum(fail) / fail_count
    print(f"不及格学生的人数为{fail_count}人，平均成绩为{fail_avg:.1f}分")
else:
    print("没有不及格的学生")


excellent_count = len(excellent)
if excellent_count > 0:
    excellent_avg = sum(excellent) / excellent_count
    print(f"优秀学生的人数为{excellent_count}人，平均成绩为{excellent_avg:.1f}分")
else:
    print("没有优秀的学生")
```
